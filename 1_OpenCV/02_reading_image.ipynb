```python
import numpy as np
import cv2
import matplotlib.image as matimg
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# OpenCV -BGR - numpy array
img_cv = cv2.imread('./data/flying_bird.jpg')
#matplotlib -RGB - numpy array
img_mat = matimg.imread('./data/flying_bird.jpg')
#PIL - RGB - numpy array
img_pil = Image.open('./data/flying_bird.jpg')
```

```python
# OpenCV BGR 
print(img_cv)
```

```python
# matplotlib -RGB
print(img_mat)
```

```python
img_pil
```

```python
# converting PIL object to array
img_pil_arr = np.array(img_pil)
print(img_pil_arr)
```

## Display Image

```python
# display image using opencv
cv2.imshow('OpenCV BGR',img_cv)
cv2.imshow('Matplotlib RGB',img_mat)
cv2.imshow('PIL RGB',img_pil_arr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# display image using matplotlib 

plt.figure(figsize=(15,8))  
plt.subplot(1,3,1)
plt.imshow(img_cv)
plt.axis('off')
plt.title('OpenCV BGR')


plt.subplot(1,3,2)
plt.imshow(img_mat)
plt.axis('off')
plt.title('Matplotlib RGB')


plt.subplot(1,3,3)
plt.imshow(img_pil_arr)
plt.axis('off')
plt.title('PIL RGB')
plt.show()
```

## COLOR Space

- Color Conversion
- Gray Scale


```python
img_cv.shape
```

```python
# numpy slicing

b = img_cv[:,:,0]
g = img_cv[:,:,1]
r = img_cv[:,:,2]
```

```python
cv2.imshow('Blue',b)
cv2.imshow('Green',g)
cv2.imshow('Red',r) 

cv2.imshow('BGR',img_cv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# opencv split channels
b, g, r = cv2.split(img_cv)
```

```python
cv2.imshow('Blue',b)
cv2.imshow('Green',g)
cv2.imshow('Red',r) 

cv2.imshow('BGR',img_cv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# converting color

# bgr to rgb

img_cv_rgb = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_cv_rgb)
plt.axis('off')
plt.show()
```

## GrayScale

```python
img_gray = cv2.cvtColor(img_cv, cv2.COLOR_BGR2GRAY)
cv2.imshow('Gray',img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
