```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create an array in the range (0,255). 16 x16

arr = np.arange(0,256,1).reshape(16,16)   
print(arr)
```

```python
arr_8bit = arr.astype(np.uint8)
plt.imshow(arr_8bit, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# random matrix (16 x 16) with values in the range (0,255)  
arr1 = np.random.randint(0,256, size=(16,16), dtype=np.uint8)
plt.imshow(arr1, cmap='gray')
plt.axis('off')
plt.show()
```
